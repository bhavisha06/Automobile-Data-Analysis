```python
#4 which company is producing most expensive car?
#max_company=df.loc[df['price'].idxmax() ,'company' ]   
#print(max_company)

df.groupby('company')['price'].max().sort_values().tail(1)
```

```python
#5 Name the cars on basis of maximum length, highest wheelbase, maximum horsepower
df.groupby('company')[['length', 'wheel-base' , 'horsepower']].max()
```

```python
#6 company wise average price of cars and order them in decending order
df.groupby('company')['price'].mean().sort_values(ascending= False)
```

```python
#7 create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"
df['decision']= np.where((df['wheel-base']<100)&(df['length']<175),
                      "don't buy" ,
                         "buy")
```

```python
#8 Select only cars whose decision is "buy"; find average price based on company of ca
df[df['decision']=="buy"].groupby('company')['price'].mean()
```
